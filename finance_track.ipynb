```python
#finance tracker
import ipywidgets as widgets
from IPython.display import display, clear_output
import pandas as pd

# Global data storage
expenses_data = []

# Input widgets (LIKE calculator inputs)
item_input = widgets.Text(value='', placeholder='Lunch, Bus...', description='Item:')
amount_input = widgets.FloatText(value=0.0, description=' Amount:')
category_dropdown = widgets.Dropdown(
    options=['Food', 'Transport', 'Study', 'Shopping', 'Other'],
    value='Food', description='Category:'
)
add_button = widgets.Button(description='Add Expense', button_style='success')

# Output areas
result_label = widgets.Label(value='Enter expense above')
total_label = widgets.Label(value='Total: â‚¹0.00 | 0 expenses')
table_output = widgets.Output()

# Add expense function (SAME as calculator logic)
def add_expense(button):
    item = item_input.value
    amount = amount_input.value
    category = category_dropdown.value

    if item and amount > 0:
        expenses_data.append({
            'Item': item,
            'Amount': amount,
            'Category': category,
            'Date': '2025-11-29'
        })
        result_label.value = f'Added: {item} (â‚¹{amount:.2f})'
        item_input.value = ''
        amount_input.value = 0.0
        update_display()
    else:
        result_label.value = ' Enter item & amount!'

add_button.on_click(add_expense)

# Update display function
def update_display():
    total = sum(exp['Amount'] for exp in expenses_data)
    total_label.value = f'Total: â‚¹{total:.2f} | {len(expenses_data)} expenses'

    with table_output:
        clear_output()
        if expenses_data:
            df = pd.DataFrame(expenses_data)
            display(df)
        else:
            print('No expenses yet - add one above!')

# Display everything
print("FINANCE TRACKER")
display(widgets.VBox([
    widgets.HTML("<h3>âž• Add New Expense</h3>"),
    item_input, amount_input, category_dropdown, add_button,
    result_label, total_label,
    widgets.HTML("<h3>ðŸ“‹ Expense History</h3>"),
    table_output
]))

update_display()
```
