```python
# Hangman Game (Nov 2025)
import ipywidgets as widgets
from IPython.display import display, clear_output
import random

# Game data
words = ['PYTHON', 'HANGMAN', 'COMPUTER', 'PROGRAM', 'ALGORITHM', 'DATABASE', 'CODING']
word = random.choice(words)
guessed_letters = set()
wrong_guesses = 0
max_wrong = 6

# Widgets
letter_input = widgets.Text(value='', placeholder='A-Z', description='Guess:')
guess_button = widgets.Button(description=' Guess Letter', button_style='info')
word_display = widgets.Label(value='_ ' * len(word))
hangman_display = widgets.Label(value='Hangman: Headless')
result_label = widgets.Label(value='Guess a letter (A-Z)!')
play_again_btn = widgets.Button(description='ðŸ”„ New Game', button_style='success')

# Hangman stages
hangman_stages = [
    '-----\n|   |\n     \n     \n     \n     ',
    '-----\n|   |\n O   \n     \n     \n     ',
    '-----\n|   |\n O   \n |   \n     \n     ',
    '-----\n|   |\n O   \n/|   \n     \n     ',
    '-----\n|   |\n O   \n/|\\  \n     \n     ',
    '-----\n|   |\n O   \n/|\\  \n/    \n     ',
    '-----\n|   |\n O   \n/|\\  \n/ \\  \nYOU LOSE!'
]

def guess_letter(button):
    global wrong_guesses

    letter = letter_input.value.upper().strip()
    letter_input.value = ''

    if len(letter) != 1 or not letter.isalpha():
        result_label.value = ' Enter ONE letter (A-Z)!'
        return

    if letter in guessed_letters:
        result_label.value = f'Already guessed {letter}! Try another.'
        return

    guessed_letters.add(letter)

    if letter in word:
        # Update word display
        new_word = ''
        for i, w in enumerate(word):
            new_word += w if w == letter or w in guessed_letters else '_ '
        word_display.value = new_word
        result_label.value = f' {letter} is in the word!'

        # Check win
        if all(w in guessed_letters for w in word):
            result_label.value = f'ðŸŽ‰ YOU WIN! Word was {word}'
            guess_button.disabled = True
    else:
        wrong_guesses += 1
        hangman_display.value = f'Hangman:\n{hangman_stages[wrong_guesses]}'
        result_label.value = f' {letter} not in word. {max_wrong - wrong_guesses} tries left.'

        if wrong_guesses >= max_wrong:
            result_label.value = f'GAME OVER! Word was {word}'
            guess_button.disabled = True

    # Show guessed letters
    guessed_str = ', '.join(sorted(guessed_letters))
    play_again_btn.layout.visibility = 'visible' if guess_button.disabled else 'hidden'

def new_game(button):
    global word, guessed_letters, wrong_guesses
    word = random.choice(words)
    guessed_letters = set()
    wrong_guesses = 0
    guess_button.disabled = False
    play_again_btn.layout.visibility = 'hidden'
    word_display.value = '_ ' * len(word)
    hangman_display.value = 'Hangman: Headless'
    result_label.value = 'New game! Guess a letter (A-Z)!'

guess_button.on_click(guess_letter)
play_again_btn.on_click(new_game)

# Display game
print("HANGMAN GAME")
display(widgets.VBox([
    widgets.HTML("<h2>ðŸ”¤ Guess the Word!</h2>"),
    word_display, hangman_display,
    widgets.HBox([letter_input, guess_button]),
    result_label,
    widgets.HTML("<h3>Guessed: None</h3>"),
    play_again_btn
]))
```
