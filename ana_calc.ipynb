```python
# GUI Calculator (Nov 2025)
from IPython.display import HTML, display
import math

html_code = """
<div style="font-family: Arial; padding: 20px; border: 2px solid #4CAF50; border-radius: 10px; max-width: 300px; margin: 20px;">
<h3>Simple Calculator</h3>
<input type="number" id="num1" placeholder="Number 1" style="width:100%; padding:10px; margin:5px; font-size:16px;">
<input type="number" id="num2" placeholder="Number 2" style="width:100%; padding:10px; margin:5px; font-size:16px;">
<select id="op" style="width:100%; padding:10px; margin:5px; font-size:16px;">
<option>+</option><option>-</option><option>*</option><option>/</option>
</select>
<button onclick="calculate()" style="width:100%; padding:15px; background:#4CAF50; color:white; border:none; font-size:18px; border-radius:5px; cursor:pointer;">Calculate</button>
<div id="result" style="padding:15px; background:#00000; margin:10px 0; font-size:20px; font-weight:bold;">Result: Enter numbers above</div>
</div>

<script>
function calculate() {
    let n1 = parseFloat(document.getElementById('num1').value) || 0;
    let n2 = parseFloat(document.getElementById('num2').value) || 0;
    let op = document.getElementById('op').value;
    let result;

    if (op == '+') result = n1 + n2;
    else if (op == '-') result = n1 - n2;
    else if (op == '*') result = n1 * n2;
    else if (op == '/') result = n2 != 0 ? n1 / n2 : 'Error';

    document.getElementById('result').innerHTML = `Result: ${result}`;
}
</script>
"""

display(HTML(html_code))
```
